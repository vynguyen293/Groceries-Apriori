```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

# Load Data

```python
data = pd.read_excel('ItemList.xlsx', header=None, engine='openpyxl')
```

```python
data.shape
```

```python
data.info()
```

```python
data
```

# Data Pre-processing

```python
dataset = []
for i in range(0, data.shape[0]):
    dataset.append([str(data.values[i,j]) for j in range(0, data.shape[1])])
```

```python
dataset
```

```python
te = TransactionEncoder()
te.fit(dataset)
te_ary = te.transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
df.info()
```

```python
df = df.drop(['nan'], axis=1)
```

```python
df.shape
```

# Use APRIORI

**Choose support = 0.001**

```python
frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
frequent_itemsets
```

**Choose confidence = 0.005**

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequent_itemsets, metric='confidence', min_threshold=0.005)
```

**Choose lift = 1.4**

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.4)
rules
```

```python
rules.info()
```

**Does the result have "sausage"?**

```python
for row in rules.iterrows():
    if 'sausage' in row[1][0]:
        print(row)
```

==> The result has "sausage". There are some items which are bought with "sausage" include:
- beverages

- whole milk
- whole milk, yogurt

```python
support = rules['support'].values
confidence = rules['confidence'].values
lift = rules['lift'].values
```

```python
plt.figure(figsize=(8,5))
plt.scatter(support, confidence, s=lift*100, alpha=0.8, c=lift)
plt.xlabel('support')
plt.ylabel('confidence');
```

```python
result = df.apply(pd.value_counts).fillna(0)
result
```

```python
df_true = result.iloc[1,:]
df_true
```

**Visualize 15 best selling products**

```python
x = df_true.sort_values(ascending=False)
x
```

```python
most_products = x[:15]
most_products
```

```python
pos = np.arange(len(most_products.values))
plt.figure(figsize=(10,6))
plt.bar(pos, most_products.values, align='center')
plt.xticks(pos, most_products.keys(), rotation='vertical')
plt.ylabel('Buying Times')
plt.xlabel('Names')
plt.title('15 best selling products');
```

15 best selling products:
**whole milk, other vegetables, rolls/buns, soda, yogurt, root vegetables, tropical fruit, bottled water, sausage, citrus fruit, pastry, pip fruit, shopping bags, canned beer, bottled beer**
